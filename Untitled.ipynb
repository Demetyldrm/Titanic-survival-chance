```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
import matplotlib.pyplot as plt
# Assuming you have a DataFrame named 'df' with columns 'Age' and 'Survived'

# Drop rows with missing age values
df_cleaned = df.dropna(subset=['Age'])

# Group data by Age and calculate the survival rate
age_survival_rate = df_cleaned.groupby('Age')['Survived'].mean()

# Create the chart
plt.figure(figsize=(10, 6))
plt.plot(age_survival_rate.index, age_survival_rate.values, marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age')
plt.grid(True)
plt.show()
```

```python
# Assuming you have a DataFrame named 'df' with columns 'Pclass' and 'Survived'

# Group data by Pclass and calculate the survival rate
pclass_survival_rate = df.groupby('Pclass')['Survived'].mean()

# Create the chart
plt.figure(figsize=(6, 4))
plt.bar(pclass_survival_rate.index, pclass_survival_rate.values, color='b')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.grid(True)
plt.show()
```

```python
# Assuming you have a DataFrame named 'df' with columns 'Age' and 'Survived'

# Drop rows with missing age values
df_cleaned = df.dropna(subset=['Age'])

# Create age groups
bins = [0, 18, 30, 50, float('inf')]
labels = ['Children', 'Young Adults', 'Middle-Aged', 'Elderly']
df_cleaned['AgeGroup'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels, right=False)

# Group data by AgeGroup and Survival and calculate the counts
age_group_survival_count = df_cleaned.groupby(['AgeGroup', 'Survived']).size().unstack()

# Create the chart
plt.figure(figsize=(10, 6))
age_group_survival_count.plot(kind='bar', stacked=True, color=['r', 'g'])
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Survival Count by Age Group')
plt.grid(True)
plt.legend(['Not Survived', 'Survived'])
plt.show()

# This data shows that each age group has similar survival rates.
```

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
df = pd.read_csv('train.csv')
df.drop(['Age','PassengerId','Name','Ticket','Cabin'], axis=1, inplace= True)
df=pd.get_dummies(df,columns = ['Sex','Embarked'])
X_train = df.drop('Survived', axis=1)
y_train = df['Survived']

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

X_train
y_train


```python
X=X_train
y=y_train
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred=logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
# Function to get input from the user and predict the status
def predict_status():
    Pclass = int(input("Enter passenger class (1, 2, or 3): "))
    Sex = input("Enter passenger gender (male or female): ")
    SibSp = int(input("Enter SibSp"))
    Parch = int(input("Enter Parch"))
    Fare = float(input("Enter passenger fare: "))
    Embarked = input("Enter embark status (C,Q,S)")
    
    
    # Preprocess the input data
    input_data = pd.DataFrame({
        'Pclass': [Pclass],
        'Sex': [Sex],
        'SibSp':[SibSp],
        'Parch':[Parch],
        'Fare': [Fare],
        'Embarked':[Embarked]
    })
    
    # Encode categorical features
    input_data = pd.get_dummies(input_data, columns=['Sex','Embarked'])
    
    # Make the prediction
    prediction = logreg.predict(input_data)[0]
    
    if prediction == 0:
        print("Not Survived")
    else:
        print("Survived")
    
predict_status()
```
